Problem statement: To build a CNN based model which can accurately detect melanoma. Melanoma is a type of cancer that can be deadly if not detected early. It accounts for 75% of skin cancer deaths. A solution which can evaluate images and alert the dermatologists about the presence of melanoma has the potential to reduce a lot of manual effort needed in diagnosis.

This uses a dataset of about 2357 images of skin cancer types. The dataset contains 9 sub-directories in each train and test subdirectories. The 9 sub-directories contains the images of 9 skin cancer types respectively.

````python
# Defining the path for train and test images
## Update the paths of the train and test dataset
data_dir_train = pathlib.Path("<Path to the train set>")
data_dir_test = pathlib.Path("<Path to the test set>")
````

````python
image_count_train = len(list(data_dir_train.glob('*/*.jpg')))
print(image_count_train)
image_count_test = len(list(data_dir_test.glob('*/*.jpg')))
print(image_count_test)
````

### Load using keras.preprocessing

Let's load these images off disk using the helpful image_dataset_from_directory utility.

### Create a dataset

Define some parameters for the loader:

````python
batch_size = 32
img_height = 180
img_width = 180
````

Use 80% of the images for training, and 20% for validation.

````python
## Write your train dataset here
## Note use seed=123 while creating your dataset using tf.keras.preprocessing.image_dataset_from_directory
## Note, make sure your resize your images to the size img_height*img_width, while writting the dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(directory=data_dir_train, image_size=(180, 180), seed=123)
````

````python
## Write your validation dataset here
## Note use seed=123 while creating your dataset using tf.keras.preprocessing.image_dataset_from_directory
## Note, make sure your resize your images to the size img_height*img_width, while writting the dataset
val_ds = tf.keras.preprocessing.image_dataset_from_directory(directory=data_dir_test, image_size=(180, 180), seed=123)
````

````python
# List out all the classes of skin cancer and store them in a list. 
# You can find the class names in the class_names attribute on these datasets. 
# These correspond to the directory names in alphabetical order.
class_names = train_ds.class_names
print(class_names)
````

### Visualize the data
#### create a code to visualize one instance of all the nine classes present in the dataset

````python
import matplotlib.pyplot as plt

### Training or validation data to visualize

# train_ds and val_ds are object of classe tf.data.Dataset -> It is an iterable

# Create dictionary of the labels 
list_label = ['actinic keratosis', 'basal cell carcinoma', 'dermatofibroma', 'melanoma', 'nevus', 'pigmented benign keratosis', 'seborrheic keratosis', 'squamous cell carcinoma', 'vascular lesion']
label_dict = dict(zip([x for x in range(9)], list_label))


batch = None
check_label = 0

# Batch is a tuple of (batch_tensor, list_labels)
for batch in val_ds.as_numpy_iterator():
  for idx in range(batch[0].shape[0]):
    
    # Get single image from the batch and corresponding label
    img = batch[0][idx,:,:,:].astype('uint8')
    label = batch[1][idx]
 
    if check_label == label:
      print(f"The label of the image : {label_dict[label]}")
      print("*"*60)
      plt.imshow(img)
      plt.show()

      # update the label flag counter, to look for the next label
      if check_label <9:
        check_label += 1
      else:
        break

    
````

The `image_batch` is a tensor of the shape `(32, 180, 180, 3)`. This is a batch of 32 images of shape `180x180x3` (the last dimension refers to color channels RGB). The `label_batch` is a tensor of the shape `(32,)`, these are corresponding labels to the 32 images.

`Dataset.cache()` keeps the images in memory after they're loaded off disk during the first epoch.

`Dataset.prefetch()` overlaps data preprocessing and model execution while training.

````python
AUTOTUNE = tf.data.experimental.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
````

### Create the model
#### Create a CNN model, which can accurately detect 9 classes present in the dataset. Use ```layers.experimental.preprocessing.Rescaling``` to normalize pixel values between (0,1). The RGB channel values are in the `[0, 255]` range. This is not ideal for a neural network. Here, it is good to standardize values to be in the `[0, 1]`

````python
# Instantiate keras models.sequential class
model = Sequential()

# Define the input layer for the model
model.add(layers.InputLayer(input_shape=(180, 180, 3)))

# Normalize the input to the scale between 0 and 1
model.add(layers.experimental.preprocessing.Rescaling(1.0/255))

# Layer 1 - 32 filters of 3x3 size with relu activation 
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D((2, 2)))

# Layer 2 - 64 filters of 3x3x32 size with relu activation
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=2))

# Layer 3 - 128 filters of 3x3x64 size with relu activation
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=2))

# Layer 4 - another layer with 128 filter with 3x3x128 size with relu activation
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(128, (3, 3), activation='relu', 
                        strides=(2,2)))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=2))

# Flatten the tensor output from the layer 3
model.add(layers.Flatten())

# Defining 2 fully connencted layers
model.add(layers.Dense(units=2048,activation="relu"))
model.add(layers.Dense(units=2048,activation="relu"))

# Softmax layer as the output probability head 
model.add(layers.Dense(units=9, activation="softmax"))
````

### Compile the model
Choose an appropirate optimiser and loss function for model training 

````python
### choose an appropirate optimiser and loss function
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

````python
# View the summary of all layers
model.summary()
````

### Train the model

````python
epochs = 20
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
````

### Visualizing training results

````python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
````

###  Findings

* The training and the validation acccurcay start at a delta of 0.25 approximately, which is visually comprehended too. The Training accuracy shows a monotonically increasing nature and a small down hill in the end epoch, which points to the fact, that the model might have had saturate on the learning capacity at the time and further training wouldn't have helped.
The validation accuracy is has sharp bumps going up and down and in the end going towards the down hill trend, which means the model's accuracy for validation set is not par with training accuracy. This clearly becomes the case of overfitting.

* The loss value trend, too show the same trend, the training and the validation losses start around the loss value of 2.0 and 2.75, but diverge from that point onwards, which clearly means it is a case of model overfit.

````python
# After you have analysed the model fit history for presence of underfit or overfit, choose an appropriate data augumentation strategy. 

"""
The model might be overfiiting over the orientation of all the images. We can try to augement the dataset by performing 
1. rotation upto 45 degrees
2. horizontal flip over each of the class of the datapoints 
3. width shift 
4. height shift
5. zoom_range

All of the these implementation helps create a different spatial view of the cancer.
"""

# Creating generator objects for training and test sets
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
                                                                horizontal_flip=True, 
                                                                rotation_range=45,
                                                                width_shift_range=.15,
                                                                height_shift_range=.15,
                                                                zoom_range=0.5
                                                                )

test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
                                                                horizontal_flip=True, 
                                                                rotation_range=45,
                                                                width_shift_range=.15,
                                                                height_shift_range=.15,
                                                                zoom_range=0.5
                                                                )

# Traversing through the train dataset folder and fetching images using the
# Generator object.
train_gen = train_datagen.flow_from_directory(data_dir_train,
        target_size=(180, 180),  
        batch_size=batch_size,
        class_mode="sparse",
        classes = list_label
        )

val_gen = train_datagen.flow_from_directory(data_dir_test, 
                                            target_size=(180, 180),
                                            batch_size=batch_size,
                                            class_mode="sparse",
                                            classes = list_label
                                            )


````

````python
# Visualize how your augmentation strategy works for one instance of training image.

# For each of the augmentation done, we are creating a visualization for one image
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    print("Left to Right different augmentations on one generator image")
    for img, ax in zip( images_arr, axes):
        ax.imshow(img.astype('uint8'))
    plt.tight_layout()
    plt.show()
    
    
augmented_images = [train_gen[0][0][0] for i in range(5)]
plotImages(augmented_images)
````

### Create the model, compile and train the model


````python
## You can use Dropout layer if there is an evidence of overfitting in your findings

"""
We found the evidence from the training accuracy and loss plots that there is 
overfitting observed. There is would be interest to try and figure out how the 
drop-out can work to mitigate the overfitting.

"""

DROPOUT_PROB = 0.2

# Instantiate keras models.sequential class
model = Sequential()

# Define the input layer for the model
model.add(layers.InputLayer(input_shape=(180, 180, 3)))

# Normalize the input to the scale between 0 and 1
model.add(layers.experimental.preprocessing.Rescaling(1.0/255))

# Layer 1 - 32 filters of 3x3 size with relu activation 
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D((2, 2)))
model.add(layers.Dropout(DROPOUT_PROB))

# Layer 2 - 64 filters of 3x3x32 size with relu activation
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=2))
model.add(layers.Dropout(DROPOUT_PROB))

# Layer 3 - 128 filters of 3x3x64 size with relu activation
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=2))
model.add(layers.Dropout(DROPOUT_PROB))

# Layer 4 - another layer with 128 filter with 3x3x128 size with relu activation
# Max pool operation with 2x2 field of view
# Followed by Batch Normalization
model.add(layers.Conv2D(128, (3, 3), activation='relu', 
                        strides=(2,2)))
model.add(layers.BatchNormalization())
model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=2))
model.add(layers.Dropout(DROPOUT_PROB))

# Flatten the tensor output from the layer 3
model.add(layers.Flatten())

# Defining 2 fully connencted layers
model.add(layers.Dense(units=2048,activation="relu"))
model.add(layers.Dropout(DROPOUT_PROB))
model.add(layers.Dense(units=2048,activation="relu"))

# Softmax layer as the output probability head 
model.add(layers.Dense(units=9, activation="softmax"))



````

### Compiling the model

````python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

### Training the model

````python
epochs = 20
history = model.fit_generator(
        train_gen,
        epochs=epochs,
        validation_data=val_gen)

````

### Visualizing the results

````python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
````

#### Findings

We see the gap between the validation loss and accuracy w.r.t the training loss and accuracy have reduced, i.e. the model is less overfitting to the training data compared to the previous run where the training accuracy was touching a score above 90% and validation score was still reaching till 30%. In this case, we see that, the training accuracy is reaching about 55% and the validation accuracy is hitting the ceiling of 38%, which points to the fact that, the overfitting has been reduced, but there is still a scope of improvement.

#### Findings
#### Many times real life datasets can have class imbalance, one class can have proportionately higher number of samples compared to the others. Class imbalance can have a detrimental effect on the final model quality. Hence as a sanity check it becomes important to check what is the distribution of classes in the data.

````python
import os
import pandas as pd
import matplotlib.pyplot as plt


# Train and Test Directory paths
train_dir = "/content/gdrive/MyDrive/Colab Notebooks/dataset/Train"
test_dir = "/content/gdrive/MyDrive/Colab Notebooks/dataset/Test"

train_set_len = 2239
test_set_len = 118

def find_dist(dir_name):
  
  # Distribution dict
  dis_dict = dict(zip(list_label, [0 for _ in range(9)]))

  for class_dir in os.listdir(dir_name):

    dis_dict[class_dir] = []
    
    # Total Count for each ckass
    samples = len(os.listdir(os.path.join(dir_name, class_dir)))
    dis_dict[class_dir].append(samples)
    
    # Calculate the proportion for each class
    if train_dir.split('/')[-1] == "Train":
      dis_dict[class_dir].append(np.round((samples/train_set_len)*100, 2))
    else:
      dis_dict[class_dir].append(np.round((samples/test_set_len)*100, 2))
  
  df = pd.DataFrame({'Labels' : list_label, 'Count': [val[0] for key, val in dis_dict.items()]})
  df['Count'].plot(kind='bar', rot=0)
  plt.show()
  print(dis_dict)

print("*"*50)
print("Training Dataset Distribution")
find_dist(train_dir)
print("*"*50)
print("Test Dataset Distribution")
find_dist(test_dir)

````
